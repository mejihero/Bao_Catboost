```python
from sklearn.datasets import load_boston
import pandas as pd

df_boston = load_boston()
df = pd.DataFrame(df_boston.data, columns = df_boston.feature_names)

df.head()
```

```python
df['MEDV'] = df_boston.target

df['LSTAT2'] = df['LSTAT']**2

df.head()
```

```python
df.isnull().sum()/df.shape[0]
```
